```python
#Courant Institute of Mathematical Scienes
#code written by Michael Lukiman
#any plagiarism will get you wrecked both directly and indirectly 
#because I spent more time on this than it seems

upper_bound = 50

trans = []
runtimes = []

for m in range(1,upper_bound):
    array = []
    for n in reversed(range(0,m)):
        array.append(n)
    print(array)
        

    transpose_count = 0

    for x in range(0,len(array)-1):
        if array[x]>array[x+1]:
            transpose_count = transpose_count + 1

    runtime = 0

    for j in range(1,len(array)):
        key = array[j]
        #print('key: '+str(key))
        i = j - 1
        while array[i]>key and i>-1: 
            runtime = runtime + 1
            array[i+1]=array[i]
            array[i]=key
            i=i-1
    print('Trans: '+str(transpose_count))
    print('Runtime: ' +str(runtime))
    print('----')
    trans.append(transpose_count)
    runtimes.append(runtime)
   
print(trans)
print(runtimes)

```
