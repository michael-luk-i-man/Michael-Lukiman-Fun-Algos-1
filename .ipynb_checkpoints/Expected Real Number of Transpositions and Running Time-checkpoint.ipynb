```python
#Courant Institute of Mathematical Scienes
#code written by Michael Lukiman
#any plagiarism will get you wrecked both directly and indirectly 
#because I spent more time on this than it seems

import random
import numpy

upper_bound = 50



for m in range(1,upper_bound): #this is the array length that gets larger, defined by the upper bound
    trans = []
    runtimes = []
    for x in range(1,100): #we will take 100 samples of each randomized array of length n
        array = []
        for n in reversed(range(1,m)): # make the array
            array.append(n)
        random.shuffle(array) # shake it up



        transpose_count = 0 

        for x in range(0,len(array)-1): # count transposes
            if array[x]>array[x+1]:
                transpose_count = transpose_count + 1

        runtime = 0

        for j in range(1,len(array)): 
            key = array[j]
            #print('key: '+str(key))
            i = j - 1
            while array[i]>key and i>-1: 
                runtime = runtime + 1
                array[i+1]=array[i]
                array[i]=key
                i=i-1
        trans.append(transpose_count)
        runtimes.append(runtime)
    print('For n = ' + str(m-1) + ': ' + str(numpy.mean(trans)) + ' | ' + str(numpy.mean(runtimes))) # Present the data
    print('-----')

```
